<h4> declaration 

```python
import time                                         # calculer le temps
import pandas as pd                              # pandas et matplotlib pour representation graphique 
import matplotlib.pyplot as plt
import math                                     # math pour calculer la racine carr√©

Numbers_premier=[1000003 , 2000003 , 4000037 , 8000009 , 16000057 , 32000011 , 64000031 , 128000003 , 256000001  , 512000009 , 1024000009 , 2048000011]

```

## Algorithm 1

nombre d'iteration N-2 , complexity : O(N)

<h4> implementation de la fonction <h4>

```python

def estpremier( x ):
    if(x<2):
        return False
    else:
        for i in range(2,x):
            if x % i == 0:
                return False 
            
        
        return True
        
        
```

<h4> Execution de la fonction sur le tableau des nombres premiers<h4>

```python
        
temps_exec=[]        
    
for x in Numbers_premier:
    
    debut= time.time()
    
    estpremier(x)                               # pour chaque nombre on calcul le temps avant et apres execution de la fonction 
    
    fin = time.time()
    
    temps_exec.append(fin-debut)
    
    
  
```

<h4> Resultat de temps de calcul pour chaque nombre et representation graphique<h4>

```python

tableau= pd.DataFrame(zip(Numbers_premier , temps_exec) , columns= [ 'number' , 'temps(s) algo_1'])
print(tableau)
```

```python
plt.figure(figsize=(8,6))
plt.plot(tableau['number'],tableau['temps(s) algo_1'] , '.-')
plt.xlabel("number")
plt.ylabel("temps(s)")
plt.savefig("algo1_graph.png")

plt.show()
```

## Algorithme 2 
Nombre d'it√©ration (N-2)/2 , complexity : O(N)

#### Implementation de la fonction

```python

def estpremier_algo2( x ):
    
    if(x<2):
        return False
    else:
        end = int(x/2)
        for i in range(2,end):
            if x % i == 0:
                return False 
                
            
        return True
            
        
```

#### Application de l'algorithme sur le tableu des nombres premiers

```python
temps_exec2=[]
for x in Numbers_premier:
    
    debut= time.time()
    
    estpremier_algo2(x)
    
    fin = time.time()
    
    temps_exec2.append(fin-debut)
    
    
```

#### Resultat de temps de calcul pour chaque nombre et representation graphique

```python

tableau2= pd.DataFrame(zip(Numbers_premier , temps_exec2) , columns= [ 'number' , 'temps(s) algo 2'] , )
print(tableau2)
```

```python
plt.figure(figsize=(8,6))
plt.plot(tableau2['number'],tableau2['temps(s) algo 2'] , '.-')
plt.xlabel("number")
plt.ylabel("temps(s)")
plt.savefig("algo2_graph.png")
plt.show()
```

## Algorithme 3 
Parcourir jusqua ‚àöN-2 , complexity o(‚àöN)


#### Implementation de la fonction

```python

def estpremier_algo3( x ):
    
    if (x<2):
        return False
    else:
        end2 = int(math.sqrt(x))
        for i in range(2,end2):
            if x % i == 0:
                return False 
                
            
        return True
```

#### Application de l'algorithme sur le tableu des nombres premiers

```python
temps_exec3=[]
for x in Numbers_premier:
    
    debut= time.time()
    
    estpremier_algo3(x)
    
    fin = time.time()
    
    temps_exec3.append(fin-debut)
    
```

#### Resultat de temps de calcul pour chaque nombre et representation graphique

```python

tableau3= pd.DataFrame(zip(Numbers_premier , temps_exec3) , columns= [ 'number' , 'temps(s) algo 3'])
print(tableau3)
```

## Algorithme 4

Nombre d'it√©rations : ‚àöN-3 / 2 , complexity O(‚àöN)

#### Implementation de la fonction

```python
    
    
def estpremier_algo4( x ):
    if x<2:
        return False
    else:
        
        if x % 2==0  and x!=2:
            return False
        else:
            end=int(math.sqrt(x))+1
            for i in range(3,end,2):
            
                if x % i == 0:
                    return False 
                    
                
            return True
```

#### Application de l'algorithme sur le tableu des nombres premiers

```python
temps_exec4=[]
for x in Numbers_premier:
    
    debut= time.time()
    
    estpremier_algo4(x)
    
    fin = time.time()
    
    temps_exec4.append(fin-debut)
    
```

#### Resultat de temps de calcul pour chaque nombre et representation graphique

```python

tableau4= pd.DataFrame(zip(Numbers_premier , temps_exec4) , columns= [ 'number' , 'temps(s) algo 4'])
print(tableau4)
```

```python
plt.figure(figsize=(8,6))
plt.plot(tableau4['number'],tableau4['temps(s) algo 4'] , '.-')
plt.xlabel("number")
plt.ylabel("temps(s)")
plt.savefig("algo4_graph.png")
plt.show()

```
