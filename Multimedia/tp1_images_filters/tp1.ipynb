```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

```

```python
img = cv2.imread(r"Sami__jiro_horikoshi_bda37552-7194-48a8-aae9-b13500d8fd3e.png")
print(type(img))
```

```python
height , width , channels = img.shape

print (f'height: {height} / width: {width} / channels: {channels}')

B , G , R = img[300,300]

print (f'blue: {B} / green: {G} / red: {R}')

cv2.imshow("image" , img)
cv2.waitKey(0)

cv2.destroyAllWindows()

```

```python
resized  = cv2.resize(img, (int(img.shape[1]/2) ,int( img.shape[0]/2)))
print(type(int(img.shape[1]/2)))

cv2.imshow("image" , resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

#...
# Convert from BGR to YCrCb color space: Y is the Luminance component, Cb and Cr
# chrominance components
img_ycrcb = cv2.cvtColor(resized, cv2.COLOR_BGR2YCrCb)
cv2.imwrite("ycrcb_random.png", img_ycrcb)
# Convert from BGR to HSV color space
img_hsv = cv2.cvtColor(resized, cv2.COLOR_BGR2HSV)
cv2.imwrite("hsv_random.png", img_hsv)
# Convert an RGB image to grayscale
gray_img = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
# concatenate image Horizontally
images = np.concatenate((cv2.cvtColor(gray_img, cv2.COLOR_GRAY2BGR), img_ycrcb,
img_hsv), axis=1)
cv2.imshow('Gray, YCrCb and HSV images', images)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
B_img, G_img, R_img = cv2.split(img)
RGB_img = cv2.merge([R_img, G_img, B_img])
plt.imshow(G_img , cmap="gray")
plt.axis('off')
plt.title('Image')
plt.show()
```

```python
figure, plots = plt.subplots(nrows=1, ncols=3)
for i, plot in enumerate(plots):
    plotimg = np.zeros(RGB_img.shape, dtype='uint8')
    plotimg[:, :, i] = RGB_img[:, :, i]
    plot.imshow(plotimg)
    plot.axis('off')
plt.show()
```

## Binarisation : 

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# concatenate image Horizontally

cv2.imshow("CCCCCCCCCCCCCCCCCC" ,  gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()


threshold = 127
_, img_bw = cv2.threshold(gray_img, threshold, 255, cv2.THRESH_BINARY)
threshold, img_bw_otsu = cv2.threshold(gray_img, threshold, 255, cv2.THRESH_OTSU)
print (f"Otsu threshold {threshold}")
figure, plots = plt.subplots(1, 2)
plots[0].imshow(img_bw, cmap='gray')
plots[0].axis('off')
plots[0].set_title('Binarized Image')
plots[1].imshow(img_bw_otsu, cmap='gray')
plots[1].axis('off')
plots[1].set_title('Otsu Binarized Image')
plt.show()
```
